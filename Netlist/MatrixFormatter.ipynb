```python
# Open the file in read mode
file_path = 'yourScatteringMatrixFile.txt'
file = open(file_path, 'r')

# Read the contents into a variable
txt = file.read()

# Close the file
file.close()
```

```python
# A bit of formatting to adapt to Python syntax
txt = txt.replace('{','[').replace('}',']').replace(' ','').replace('==', '=')
```

```python
# Open the file in write mode
file_path = 'yourScatteringMatrixPy.txt'
file = open(file_path, 'w')

# Write the string to the file
file.write(txt)

# Close the file
file.close()
```
